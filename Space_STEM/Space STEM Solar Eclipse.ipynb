#### Looking at the graphs and the data, what do the trends tell you? 

```python
import matplotlib.pyplot as plt

dftemp = pd.DataFrame(df,columns=['Time','Temp deg C'])
dftemp.plot(x ='Time', y='Temp deg C', kind = 'line',figsize=(12,2))
plt.show()

dflux = pd.DataFrame(df,columns=['Time','Lux'])
dflux.plot(x ='Time', y='Lux', kind = 'line',figsize=(12,2))
plt.show()

dfRH = pd.DataFrame(df,columns=['Time','Rel. Humidity (%)'])
dfRH.plot(x ='Time', y='Rel. Humidity (%)', kind = 'line',figsize=(12,2))
plt.show()

dfpressure = pd.DataFrame(df,columns=['Time','Pressure (Pa)'])
dfpressure.plot(x ='Time', y='Pressure (Pa)', kind = 'line',figsize=(12,2))
plt.show()

```

As the minutes went by during the eclipse, the temp decrease gradually with its lowest at 33.21 celsius before sleepily raising

the Lux decrease and begin raising at 14:27

the relative humidity raise from 50.55 to 65.84

the Pressure decreased also from 100348.031 to 100282.292

#### Correlation

```python
df.columns
```

```python
column_1 = df["Lux"]
column_2 = df["Temp deg C"]
correlation = column_1.corr(column_2)


if correlation > 0:
    print ("The relationship is positive correlated")
elif correlation < 0:
    print ("The relationship is negative correlated")
else:
    print ("No relationship between variables")

print (correlation)    
```

```python
column_1 = df["Lux"]
column_2 = df["Rel. Humidity (%)"]
correlation = column_1.corr(column_2)


if correlation > 0:
    print ("The relationship is positive correlated")
elif correlation < 0:
    print ("The relationship is negative correlated")
else:
    print ("No relationship between variables")

print (correlation) 
```

```python
column_1 = df["Lux"]
column_2 = df["Pressure (Pa)"]
correlation = column_1.corr(column_2)


if correlation > 0:
    print ("The relationship is positive correlated")
elif correlation < 0:
    print ("The relationship is negative correlated")
else:
    print ("No relationship between variables")

print (correlation) 
```
